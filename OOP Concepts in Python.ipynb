# Inheritance:

```python
class Vehicle:
    def __init__(self,name,color,cost):
        self.name = name
        self.color = color
        self.cost = cost
    def show_vehicle_details(self):
        print("I am Vehicle")
        print("Name of vehicle is ",self.name)
        print("Color of vehicle is ",self.color)
        print("Cost of vehicle is ",self.cost)
```

```python
v = Vehicle('Bus','Blue','5000$')
```

```python
v.show_vehicle_details()
```

```python
class Car(Vehicle):
    def show_car_details(self):
        print("I am Car")
```

```python
c = Car('Mercedes','Black','8000$')
```

```python
c.show_car_details()
```

```python
c.show_vehicle_details()
```

# Over_riding of method:

```python
class Vehicle:
    def __init__(self,name,color,cost):
        self.name = name
        self.color = color
        self.cost = cost
    def show_vehicle_details(self):
        print("I am Vehicle")
        print("Name of vehicle is ",self.name)
        print("Color of vehicle is ",self.color)
        print("Cost of vehicle is ",self.cost)
```

```python
v = Vehicle('Bus','White','6000$')
```

```python
class Car(Vehicle):
    def __init__(self,name,color,cost,tyre,hp):
        super().__init__(name,color,cost)
        self.tyre = tyre
        self.hp = hp
    def show_vehicle_details(self):     # overriding the method
        super().show_vehicle_details()    # call the parent class method
        print("Number of tyres are ",self.tyre)
        print("Number of horse powers are ",self.hp)
        
```

```python
c1 = Car('Toyota','Black','7000$',4,250)
```

```python
c1.show_vehicle_details()
```

If you write "show_car_details" method in child class then it is not override.It will be a new method.Because overriding means "name of method is same that is used in parent class".

# Multiple inheritance: 

```python
class Parent1:
    def assign_string1(self,str1):
        self.str1 = str1
    def show_string1(self):
        return self.str1
```

```python
class Parent2:
    def assign_string2(self,str2):
        self.str2 = str2
    def show_string2(self):
        return self.str2
```

```python
class Child(Parent1,Parent2):
    def assign_string3(self,str3):
        self.str3 = str3
    def show_string3(self):
        return self.str3
    
```

```python
my_child = Child() 
```

```python
my_child.assign_string1("I am string of parent 1")
```

```python
my_child.assign_string2("I am string of parent 2")
```

```python
my_child.assign_string3("I am string of child")
```

```python
my_child.show_string1()
```

```python
my_child.show_string2()
```

```python
my_child.show_string3()
```

# Multi level inheritance: 

```python
class Parent:
    def set_name(self,n):
        self.name = n
    def show_name(self):
        return self.name
```

```python
class Child(Parent):
    def set_age(self,a):
        self.age = a
    def show_age(self):
        return self.age
```

```python
class GrandChild(Child):
    def set_gender(self,g):
        self.g = g
    def show_gender(self):
        return self.g
```

```python
gc = GrandChild()
```

```python
gc.set_name('Khizar')
```

```python
gc.set_age(20)
```

```python
gc.set_gender('Male')
```

```python
gc.show_name()
```

```python
gc.show_age()
```

```python
gc.show_gender()
```
