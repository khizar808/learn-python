# FUNCTIONS:

```python
def is_prime(number):
    if number % 2!=0:
        return True
    else:
        return False
```

```python
is_prime(5)
```

```python
def is_even(number):
    if number % 2==0:
        return True
    else:
        return False
```

```python
is_even(7)
```

```python
def calculate_area(base,height):
    return 1/2*(base*height)
```

```python
a = calculate_area(20,30)
print ("Area:",a)
```

```python
def calculate_perimeter(length,width):
    return 2*(length+width)
```

```python
l = int(input("Enter the length:"))
w = int(input("Enter the width:"))
p = calculate_perimeter(l,w)
print("Perimeter:",p)
```

```python
def check_type(year):
    assert type(year)== int,'The type of year is not integer'
    print('The type of year is valid')
    print('successful')
    return True
```

```python
check_type(2015)
```

```python
check_type(2011.4)
```

```python
primes = []

for num in range(2, 101):
    is_prime = True
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        primes.append(num)

print("Prime numbers up to 100 are:")
print(primes)
```
